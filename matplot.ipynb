# MATPLOTLIB 

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,11)
print(x)
y = x**2
print(y)
```

```python
plt.plot(x,y)
plt.title("hello")
plt.xlabel("x--->")
plt.ylabel("y--->")
```

```python
plt.subplot(2,2,1)
plt.plot(y,x)
plt.subplot(2,2,2)
plt.plot(x**2,x**2)
plt.subplot(2,2,3)
plt.plot(x,x**2)
plt.subplot(2,2,4)
plt.plot(x,y)
```

# object oriented way

```python
fig = plt.figure()
print(fig)
axis1 = fig.add_axes([0.1,0.1,0.8,0.8])
axis1.plot(x,y)
axis2 = fig.add_axes([0.3,0.2,0.5,0.4])
axis2.plot(x,y)
```

# types of plot

```python
plt.scatter(x,y)
```

```python
plt.hist(x)
```

```python
fig = plt.figure()
a = fig.add_axes([0.1,0.1,0.3,0.3])
a.plot(x,y)
plt.savefig("basic_plot.png")
```
